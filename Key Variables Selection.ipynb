## Merging Dataframes

```python
#Making column names consistent in snake case and lower 
school_data.columns = school_data.columns.str.lower().str.replace(" ", "_")
overall_pop_employ_data.columns = overall_pop_employ_data.columns.str.lower()
overall_income_data.columns = overall_income_data.columns.str.lower()
comp_int_est.columns = comp_int_est.columns.str.lower()
```

## !! WARNING !!
NOTE: Need to double check merges to see if data is being dropped unintentially due to which data frame we started with. 


```python
set1 = pop_data.merge(broadband_2018, left_on='id', right_on='id', how = 'outer')
set1 = set1.rename(columns = {"county_x":"county", "state_x":"state"})
set1 = set1.drop(columns = ["cfips", "county_y", "state_y"])

set2 = set1.merge(school_data, left_on='id', right_on='id', how = 'outer')
set2 = set2.rename(columns = {"county_x":"county", "state_x":"state"})
set2 = set2.drop(columns = ["county_y", "state_y"])

set3 = set2.merge(overall_income_data, left_on="id", right_on = "geo_id", how = "outer")
set3 = set3.rename(columns = {"county_x":"county", "state_x":"state"})
set3 = set3.drop(columns = ["county_y", "state_y"])
set3

set4 = set3.merge(overall_pop_employ_data, left_on = "id", right_on = "geo_id", how = "outer")
set4 = set4.rename(columns = {"county_x":"county", "state_x":"state"})
set4 = set4.drop(columns = ["county_y", "state_y"])
set4

set5 = set4.merge(comp_int_est, left_on = "id", right_on = "geo_id", how = "outer")
set5 = set5.rename(columns = {"county_x":"county", "state_x":"state"})
set5 = set5.drop(columns = ["county_y", "state_y"])

set6 = set5.merge(raw_clean_households_data, left_on = "id", right_on = "geo_id", how = "outer") #don't think it needs to be an outer join here
set6 = set6.rename(columns = {"est_total_households_x":"est_total_households_frm_income", "est_total_households_y":"est_total_households_frm_devices"})
set6 = set6.drop(columns = ["geo_id_x", "geo_id_y"])

initial_analytic_dataset = set6
```
