```python
def lookup_word(word):
    #use passed argument of a word as key lookup to get value
    lookup_dict = {'$20,000:':'20K', 'computing':'comp', 
                   'portable':'port', 'with':'w', 'AL':'alone', 'alone':'AL','or':'or', 'desktop':'DKT', '$74,999':'75k', 
                   'id':'id', 'of':'of', 'up':'U', 'estimate':'EST', 'other':'OTH', 'area':'area', 'dsl':'DSL', 
                   'subscription:':'sub', 'broadband':'BB', 'household':'hhld', 'no':'no', 'satellite':'SAT', 
                   '12':'12', 'one':'1', 'devices':'DVCs', 'error':'er', 'laptop':'LT', 'has':'', 
                   'dial':'D', 'total':'T', 'geographic':'GEO', 'months':'months', 'income':'income', 
                   '2019':'', 'without':'WO', 'smartphone':'SMP', 'types':'types', 'subscriptions':'subs', 
                   'cellular':'cell', 'dollars)':'', 'service':'service', 'past':'', 'computer':'comp_d',
                   'devices:':'', 'adjusted':'', 'than':'than', 'optic':'OPT', '$20,000':'20k',
                   'fiber':'fiber', 'name':'name', 'more:':'+', 'type':'type', 'wireless':'Wless', 
                   'an':'', 'any':'any', 'less':'less', 'as':'as', 'plan':'pln', 'such':'such',
                   'tablet':'TBL', 'the':'the', 'more':'more', '(in':'', 'internet':'INT', 'in':'in',
                   'percent':'PCT', 'inflation':'', '$75,000':'75k', 'to':'to', 'margin':'M', 
                   'device':'DVC', 'data':'', 'households':'hhlds', 'subscription':'sub', 'a':'a', 
                   'cable,':'CBL', '$74,999:':'75k'}
    ## return value based on provided key
    return lookup_dict[word]
```

```python
def build_lookup_dict(word_set):
    lookup_dict = dict.fromkeys(word_set, '0')
    return lookup_dict
    
```

```python
def clean_columns(df):
    
    df.columns = df.columns.str.lower().str.replace('!!', '_').str.replace(' ', '_').str.replace('-', '_').str.replace(':_', '_')
    #df.columns = df.columns.str.replace(r"\(.*\)", "", regex=True)
    #print(df.columns[2])
 
    lookup = {}
    #define a set outside of for loop  
    words = set()
    for x,y in enumerate(df.columns):
        # split every column into a list
        clean = y.split('_')
        
        #created our set - printed and edited locally
        [words.add(w) for w in clean]
        look_up_dict = build_lookup_dict(words)
        
        new_column_name = []
        for w in clean:
            #print(w)
            z = lookup_word(w)
            new_column_name.append(z)
        # join each element on passed charachter
    
        new_y = '_'.join(new_column_name)
        df.columns = df.columns.str.replace(y, new_y)
        #df[new_y] = df[y]
    print(look_up_dict)
    return df
        
    
cleaned_df = clean_columns(df)
```
