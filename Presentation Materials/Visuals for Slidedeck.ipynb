```python
#Packages import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_initial = pd.read_csv("initial_analytic_dataset.csv")
```

```python
df_initial.info()
```

```python
df_initial["mod_est_med_income_households"] = df_initial["est_med_income_households"]*.001

plt.figure(figsize=(10, 5))
sns.histplot(data=df_initial,
             x="mod_est_med_income_households",
             kde=True,
            color = '#4f5ee7')
plt.xlabel("Thousands of U.S. Dollars")
plt.title("Majority of households in U.S. counties have median incomes around $60K"); 
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df_initial,
             x="est_unemp_pop_ratio_16_over",
             kde=True, 
            color = '#4f5ee7')
plt.xlabel("Unemployment Rate")
plt.title("Most counties are around a 4% unemployment rate"); 
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df_initial,
             x="pct_total_households_with_desktop",
             kde=True, 
            color = '#4f5ee7')
plt.xlabel("% Households with Desktops")
plt.title("Percentage of households with desktops in counties are between 70 - 80%"); 
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df_initial,
             x="pct_total_households_with_portable",
             kde=True, 
            color = '#4f5ee7')
plt.xlabel("% Households with Portable Devices (e.g. Tablets)")
plt.title("Percentage of households with portable devices in counties are around 60%"); 
```
